# Is normally Distributed

- mpg is normally distributed (assume)

```python
from scipy import stats
import numpy as np
```

### Detect the outliers

```python
df [ np.abs(stats.zscore(df.mpg)) > 3 ]
```

### Remove Outliers

```python
df [ np.abs(stats.zscore(df.mpg)) <= 3 ]
```

# Is not normally distributed

# IQR

```python
Q1 = np.percentile(df.mpg, 25)
Q1
```

```python
Q3 = np.percentile(df.mpg, 75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit= Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

### Detect the outliers

```python
df [ (df.mpg < lower_limit) | (df.mpg > upper_limit) ]
```

### Remove the outliers

```python
df_temp = df [ (df.mpg >= lower_limit) & (df.mpg <= upper_limit) ]
```
