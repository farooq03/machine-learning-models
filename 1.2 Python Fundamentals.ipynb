bmi = weight / height ** 2

- Underweight  <18.5
- Normal weight  >= 18.5– <25
- Overweight  >= 25– <30
- Obesity  BMI of 30 or greater

```python
weight = 70
height = 1.75

bmi = weight / height ** 2

if bmi<18.5:
    print('Underweight')
elif bmi>=18.5 and bmi <25:
    print('Normal weight')
elif bmi >=25 and bmi <30:
    print('Over weight')
else:
    print('Obesity')
```
