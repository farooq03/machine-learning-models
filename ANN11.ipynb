```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('datasets/breast_cancer.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.outcome
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-1,1))
x_scale=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x_scale
```

```python
x_scale.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,random_state=10,test_size=0.2)
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(128,64,32),max_iter=300)
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
