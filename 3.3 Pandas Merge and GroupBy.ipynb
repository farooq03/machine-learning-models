```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'name': ['Arpit', 'Anick', 'Asma', 'Joseph'],
    'english': [95, 92, 94, 95]
})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'name': ['Arpit', 'Digamber', 'Asma', 'Joseph'],
    'maths': [90, 91, 92, 94]
})
```

```python
df2
```

# inner merge

```python
pd.merge(df1, df2, on='name')
```

# left merge

```python
pd.merge(df1, df2, on='name', how='left')
```

# right join

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='name', how='right')
```

# outer merge

```python
pd.merge(df1, df2, on='name', how='outer')
```

# GroupBy

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
df.cyl.unique()
```

```python
cyl_group = df.groupby('cyl')
```

```python
cyl_group.mean()
```

# which cars have more mpg automatic or manual

```python
am_group = df.groupby('am')
```

```python
am_group.mean()
```
