```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import numpy as np
```

# Line Plots

```python
x = [2017, 2018, 2019, 2020, 2021]
y = [   7,   10,   15,   30,   40]
```

```python
plt.plot(x,y)
plt.xticks(x)
plt.title('Year vs Turnover')
plt.xlabel('Year')
plt.ylabel('Turnover in lacs')
plt.show()
```

```python
x_cricket = [1, 2, 3, 4, 5, 6]
y_cricket = [10, 20, 6, 12, 12, 30]
```

```python
plt.plot(x_cricket, y_cricket, color='red')
plt.title('Overs vs Runs')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```

```python
x_math = np.arange(10)
y_math = 4*x_math + 5
plt.plot(x_math, y_math)
```

# Bar Plot

```python
plt.bar(x,y)
plt.title('Year vs Turnover')
plt.xlabel('Year')
plt.ylabel('Turnover in lacs')
plt.show()
```

```python
plt.bar(x_cricket ,y_cricket)
plt.title('Overs vs Runs')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```

# Scatter Plot

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x_cricket, y_cricket)
plt.title('Overs vs Runs')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```

# Drawing multiple plots onto a single graph

```python
x_company = [2016, 2017, 2018, 2019, 2020]
y_companyA = [12,  15,    17,   19,  22]
y_companyB = [8,   16,    25,   30,  35]
```

```python
plt.plot(x_company, y_companyA, color='green', lw=2, ls=':')
plt.plot(x_company, y_companyB, color='red', lw=2)
plt.legend(['CompanyA', 'CompanyB'])
plt.title('Comparison of growth (CompanyA & CompanyB)')
plt.xlabel('Year')
plt.ylabel('Profit in lacs')
plt.xticks(x_company)
plt.show()
```

# Pie Charts

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
profit =   [50,    75,   35,   40]
```

```python
plt.pie(profit, labels=quarters, autopct='%.1f',explode=[0, 0, 0.2, 0.0])
plt.title('Quarterly Profit')
plt.show()
```

# Subplots

```python
rcParams['figure.figsize'] = (10,10)
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,0].set_title('Line Plot')
ax[0,1].bar(x_company, y_companyA)
ax[0,1].set_title('Bar Plot')
ax[1,0].pie(profit)
ax[1,0].set_title('Pie Chart')
ax[1,1].hist(shoe_size, bins=7)
ax[1,1].set_title('Histogram')
plt.show()
```
