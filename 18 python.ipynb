# pandas merge and groupby

```python
 df1=pd.DataFrame({
     'name':['farooq','asma','anick','abdali'],
     'maths':[67,88,98,76]
 })
```

```python
df1
```

```python
 df2=pd.DataFrame({
     'name':['farooq','farah','anick','abdali'],
     'maths':[67,90,98,76]
 })
```

```python
pd.merge(df1,df2,on='name')
```

```python
pd.merge(df1,df2,on='name',how='left')
```

```python
pd.merge(df1,df2,on='name',how='right')
```

```python

pd.merge(df1,df2,on='name',how='outer')
```
